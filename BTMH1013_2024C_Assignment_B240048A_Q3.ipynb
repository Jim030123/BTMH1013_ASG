Q3 (c) Customisations                                                             
• Customise the map by adding different marker colours based on the type
of attraction (Eg: historical site, natural wonder, amusement park)

• Add a layer control to toggle different types of attractions on and off.

```python
# Import libraries
import folium
from IPython.display import display, HTML

# Create map object centered in Malaysia


# Color coding based on type
color_map = {
    "Historical Site": "blue",
    "Natural Wonder": "green",
    "Amusement Park": "red",
    "Cultural Heritage": "yellow",
    "Religious Site": "purple",
    "Beach": "orange",
    "Aquarium": "cadetblue",
    "Wildlife Sanctuary": "pink",
    "Marine Sanctuary": "darkred",
    "Garden": "darkgreen"
}

# Initialize the map centered in Malaysia
my_map = folium.Map(location=[4.210484, 101.975766], zoom_start=7)

# Layer controls dictionary
layers = {}

# Add attractions to the map
for attraction in attractions:
    attr_type = attraction["type"]
    if attr_type not in layers:
        # Create a new layer for each unique type
        layers[attr_type] = folium.FeatureGroup(name=attr_type)

    # Create a popup and marker
    popup = f"<b>{attraction['name']}</b><br>{attraction['desc']}"
    folium.Marker(
        location=[attraction["lat"], attraction["lon"]],
        popup=folium.Popup(popup, max_width=300),
        icon=folium.Icon(color=color_map.get(attr_type, "gray"))
    ).add_to(layers[attr_type])

# Add all layers to the map
for layer in layers.values():
    layer.add_to(my_map)

# Add layer control to the map
folium.LayerControl(position="topright", collapsed=True).add_to(my_map)

# Save the map to an HTML file
my_map.save("tourist_map.html")

# Display the map in a Jupyter Notebook
display(HTML(my_map._repr_html_()))
```
